## Research Question

What is the association between video game sales in Europe and the educational performance (PISA scores) among 15 year-olds in Europe between 2000 and 2015?

## Background and Prior Work

The interactive media known as video games emerged in the mid 20th century and expanded rapidly during the 1980s and 1990s with the massive rise of home consoles and personal computing. By the end of the 20th century, tens of thousands of video games have been released globally, and millions of video games sold worldwide. Video games are not just a source of entertainment, but they are also a medium which has significantly shaped culture, technological literacy and society. Iconic characters such as Mario, Pikachu, Sonic, and Steve have become globally recognized symbols, reflecting the widespread cultural influence of video games.

With the critical success and acclaim of video games, their consumption worldwide has grown dramatically and continues to rise even today. Such consumption raises questions about the possible effects of video games, considering that it has become part of daily life for many individuals on a global scale. In Europe, accessibility to video games increased greatly throughout the 20th century, such that gaming became a common form of entertainment across different age groups and socioeconomic backgrounds. The widespread availability and impact of such forms of entertainment provide an opportunity to examine how varying levels of video game consumption correspond with educational performance indicators, such as literacy rates, standardized test scores, and broader academic achievement across European populations.

This growing interest in video games and its possible effects on education has prompted researchers and scholars. Previous studies have explored many distinct ways in which video games can affect an individual particularly when it comes to their education, including aspects of education such as learning, motivation, skill development, and more. With this, and by situating it within the context of education in Europe, we can discover from previous work how historical patterns of video game consumption correspond with broader trends in academic achievement and literacy.

Prior research has examined a possible relationship between video games and academic performance in regards to student education. In a survey-based analytical study conducted on students from a charter school in Spain, Merino-Campos et al. (2023) identified that factors such as accessibility, enjoyment, and technological competence significantly influenced students’ acceptance of video games as learning tools, and that this acceptance was associated with improved academic performance in physical education contexts. This study made use of regression analysis of latent factors based on the optimization technique of partial least squares (PLS) to derive such findings. This work suggests that video game engagement can have measurable educational effects, supporting further investigation into broader relationships between video game consumption and educational outcomes. [1]

Other studies show not only the relationship of learning tools with video games, but also the social and behavioral effects that come with such media. Through examining the history of games in educational research, Squire (2003) observed that although often ignored by educators, video games have become a relevant resource for the development and improvement of cultural competence, as well as social and cognitive skills of secondary school students in the context of physical education. They draw on theories revolving around motivation, in which they conclude that deeper learning and engagement can be produced by the key elements of video games, such as challenges, feedback, and other aspects. [2]

Finally, de Wit, Ferencz, and Rumbley (2018) focus their research on international mobility in Europe at a population level. They examine international mobility as a major indicator of educational strength, ability, accessibility, and quality across Europe. This research dates to the late 20th century, which was around the same time period when video games experienced a surge in popularity and consumption. While the study does not directly examine video games, it highlights the importance of population-level indicators when evaluating educational strength across Europe. [3]

Although prior studies have explored the relationship between video games and academic outcomes, existing research has rarely investigated whether the growth of the video game industry itself is associated with changes in educational performance at a population level. This project addresses this gap by combining large-scale video game industry data with international PISA performance scores across European countries between 2003 and 2018. By analyzing trends over time, this study provides an informed, broader perspective on how changes in video game popularity may relate to educational outcomes.

## Hypothesis


Hypothesis: Increased video game sales per capita in Europe is strongly associated with lower average PISA scores among the population between 2003 and 2018.

The Programme for International Student Assessment (PISA) is an international standardized assessment conducted by the OECD that measures students’ literacy in reading, mathematics, and science, with an average score of approximately 500 across participating countries.

As stated in the background, due to the rise of video games in the mid twentieth century, countries were quick to adopt them, allowing games to become part of their culture and entertainment. More developed countries, such as Europe, were able to achieve this at a much faster rate. During that same time period in Europe, there was also a decline in educational performance as measured by the PISA Scores of 15 year-old students. We suggested that the rise of video games/video game sales is associated with lower educational performance in this age group, since video games can have varying impacts on students by taking away from study time and creating distractions. This does not explain a cause-and-effect relationship, but rather an association based on observations, which will be explored with supporting data.

#### Description 
This dataset, scraped from vgchartz.com, includes ranking, name, platform, release year, genre, publisher, and sales for games exceeding 100,000 copies. The metric used is per million copies sold (one singular video game purchased), categorized by regions: NA_Sales, EU_Sales, JP_Sales, Other_Sales, and Global_Sales (the sum of all regions). Our analysis specifically utilizes:

- Sales by Year and Genre (1980-2020): Tracking trends via a stacked area chart
- EU Region (2000-2015): Specifically examining data represented as Zone Sales

#### Concerns regarding Dataset 
- **Population Bias**: Each region contains different population numbers (e.g., North America vs. Japan). Larger populations create larger potential consumer markets, meaning sales can weigh more on market size than actual popularity.
Regulatory Variance: Regional regulations (ratings, advertising, and privacy) influence consumption. Stricter laws or total bans in certain countries effectively control the release and sale of video games.
- **Selection Bias**: Only games with over 100,000 copies sold are included. This favors larger companies and franchises, leading to an underrepresentation of indie and less popular games.

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

vg_sales = pd.read_csv('data/00-raw/Sales_by_Year_& _Genre.csv')
vg_sales.head()
```

```python
clean_vg_sales = vg_sales.dropna()
clean_vg_sales[clean_vg_sales.isnull().any(axis=1)]
```

```python
clean_vg_sales = clean_vg_sales.sort_values(by=['Zone Sales', 'Year'], ascending=[False, False])
clean_vg_sales = clean_vg_sales.reset_index(drop=True)
print(clean_vg_sales.head(10))
```

```python
clean_vg_sales = clean_vg_sales[(clean_vg_sales['Year'] >= 2000) & (clean_vg_sales['Year'] <= 2015)]

print(f"Earliest Year in Data: {clean_vg_sales['Year'].min()}")
print(f"Latest Year in Data: {clean_vg_sales['Year'].max()}")
```

```python
clean_vg_sales.shape
```

```python
clean_vg_sales
```

```python
yearly_sales = (
    clean_vg_sales
        .groupby("Year", as_index = False)["Zone Sales"]
        .sum()
)
yearly_sales = yearly_sales.rename(columns={"Zone Sales": "EU Sales"})
yearly_sales
```

The trend line and summary statistics of yearly video game sales from 2000–2015 provide insight into overall market performance during this period. The data shows that peak yearly sales occurred in 2009, reaching a maximum of 191.59 million units, which likely represents the height of the video game market within this timeframe. In contrast, the lowest recorded sales were 52.75 million units, indicating that the market experienced substantial changes in performance across the years. This variability is further supported by the standard deviation of 37.71 million, suggesting notable fluctuations in yearly sales rather than consistent growth. Additionally, the interquartile range (25%–75%) shows that most yearly sales fell between approximately 106 million and 162 million units, representing the typical performance range of the video game market between 2000 and 2015.
```python
#set range of years of 2000-2015
years = range(1, 16)

# We compute the current years sales (in Millions) minus the previous year's sales, then divide by the previous year’s sales, and multiply by 100 to get the percent change
sales_change = [42, 15, -5, 4, 15, 8, 32, 24, 8, -15, -9, -48, 8, 0, -28]

# Line Plot and Time Series of Percentage Change
plt.figure(figsize=(10, 2)) 
plt.plot(years, sales_change, marker='o', color='tab:orange', linewidth=2, label='Yearly Change')
plt.axhline(0, color='gray', linewidth=1.5, linestyle='-') 
plt.title('Year-to-Year Change in Video Game Sales', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Change in Sales (%)', fontsize=12)
plt.xticks(years) 
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

The transition into the 2010s shows a significant downturn, as sales decreased between 2010 and 2012 by approximately 15–40 million units compared to previous years. In 2013, sales experienced a noticeable increase, followed by another decline beginning in 2014 and continuing toward the end of the observed period. Overall, the changes observed from 2010 onward suggest slower growth in video game sales compared to the previous decade.

## Ethics 


Overall statement: This project is based on aggregated, publicly available databases and therefore does not contain any personally identifiable information (PII). However, it is possible that this data could be biased through collection biases resulting from differences in how consistently video game sales have been reported across countries and over time, especially in earlier time frames; additionally, the measures of educational performance are likely to be impacted by broader systemic factors that do not directly relate to video game consumption, such as the country’s overall economic development or its academic policies.
To minimize these limitations, this project will not make any causal statements about the relationship between video games and educational outcomes. 

One important ethical consideration is the potential misrepresentation and bias of the PISA dataset. The PISA dataset only includes assessments of 15-year-old students who are currently enrolled in school, meaning it does not represent all students within a country. Students who are not formally enrolled in academic institutions, due to circumstances like dropping out or being homeschooled, are not included in this sample size. As a result, the dataset may overrepresent students from more stable educational environments and underrepresent disadvantaged populations. This important limitation means that PISA scores may not fully reflect the educational performance of the entire youth population in each country.

The project will state all limitations and will present the results using graphical representations and summary statistics that will allow for transparency and show the uncertainty within the data sets. Additionally, any observed correlations should not be interpreted as evidence of individual-level effects, and no policy recommendations can be made based on the observed correlations without additional supportive evidence.

## Discussion and Conclusion


This project sought to determine whether increases in video game sales in Europe are associated with lower educational performance as measured by PISA scores. Through combining large-scale video game industry data with international PISA performance scores across European countries between 2003 and 2015, we further expanded on past research, investigating whether the growth of the video game industry itself is associated with changes in educational performance at a population level. We initially hypothesized that increased video game sales per capita in Europe are strongly associated with lower average PISA scores among the population, as video games might be distracting for students and take time away from studying. 

Graphical visualizations of the time trends showed that EU video game sales increased between 2003 and 2009 before declining by 2015, while average PISA scores showed an overall gradual decrease over the same time period. Although these trends occasionally appeared to move in opposite directions, the pattern was not consistent across all years. We then used correlation and regression analysis to further examine the relationship between the two variables. The Pearson correlation coefficient was approximately 0.24, indicating a weak positive relationship between EU video game sales and PISA scores. The OLS regression produced a coefficient of 0.0138, meaning increases in video game sales were associated with a very small increase in PISA scores. However, the relationship was not statistically significant (p = 0.695), and the model explained only 5.8% of the variation in PISA scores (R² = 0.058). Overall, these results do not support the hypothesis that higher video game sales are associated with lower PISA performance.

There are several limitations affecting the interpretation of these findings. The biggest limitation is the small sample size, as only five observations remained after merging the datasets since PISA assessments only happen once every 3 years. The PISA dataset we used also only accounts for formally enrolled students, which does not represent all 15-year-olds accurately, while the Video Game Sales dataset only includes popular games with over 100,000 sales, excluding smaller indie games. Additionally, the analysis relies solely on combined regional data, which does not reflect individual student behavior or account for other factors that influence educational outcomes, like economic conditions, educational policies, or access to technology. 

Future research could be conducted to improve this analysis by breaking the data down by country, adding more relevant variables, and increasing the number of observations over time for a more detailed and accurate analysis. Overall, the results suggest that there is no strong statistical evidence linking higher video game sales to lower educational performance among students in Europe, and that the relationship between gaming and education is more likely to be influenced by other, broader factors.
