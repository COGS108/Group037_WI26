## Background and Prior Work

The interactive media known as video games emerged in the mid 20th century and expanded rapidly during the 1980s and 1990s with the massive rise of home consoles and personal computing. By the end of the 20th century, tens of thousands of video games have been released globally, and millions of video games sold worldwide. Video games are not just a source of entertainment, but they are also a medium which has significantly shaped culture, technological literacy and society. Iconic characters such as Mario, Pikachu, Sonic, and Steve have become globally recognized symbols, reflecting the widespread cultural influence of video games.

With the critical success and acclaim of video games, their consumption worldwide has grown dramatically and continues to rise even today. Such consumption raises questions about the possible effects of video games, considering that it has become part of daily life for many individuals on a global scale. In Europe, accessibility to video games increased greatly throughout the 20th century, such that gaming became a common form of entertainment across different age groups and socioeconomic backgrounds. The widespread availability and impact of such forms of entertainment provide an opportunity to examine how varying levels of video game consumption correspond with educational performance indicators, such as literacy rates, standardized test scores, and broader academic achievement across European populations.

This growing interest in video games and its possible effects on education has prompted researchers and scholars. Previous studies have explored many distinct ways in which video games can affect an individual particularly when it comes to their education, including aspects of education such as learning, motivation, skill development, and more. With this, and by situating it within the context of education in Europe, we can discover from previous work how historical patterns of video game consumption correspond with broader trends in academic achievement and literacy.

Prior research has examined a possible relationship between video games and academic performance in regards to student education. In a survey-based analytical study conducted on students from a charter school in Spain, Merino-Campos et al. (2023) identified that factors such as accessibility, enjoyment, and technological competence significantly influenced students’ acceptance of video games as learning tools, and that this acceptance was associated with improved academic performance in physical education contexts. This study made use of regression analysis of latent factors based on the optimization technique of partial least squares (PLS) to derive such findings. This work suggests that video game engagement can have measurable educational effects, supporting further investigation into broader relationships between video game consumption and educational outcomes. [1]

Other studies show not only the relationship of learning tools with video games, but also the social and behavioral effects that come with such media. Through examining the history of games in educational research, Squire (2003) observed that although often ignored by educators, video games have become a relevant resource for the development and improvement of cultural competence, as well as social and cognitive skills of secondary school students in the context of physical education. They draw on theories revolving around motivation, in which they conclude that deeper learning and engagement can be produced by the key elements of video games, such as challenges, feedback, and other aspects. [2]

Finally, de Wit, Ferencz, and Rumbley (2018) focus their research on international mobility in Europe at a population level. They examine international mobility as a major indicator of educational strength, ability, accessibility, and quality across Europe. This research dates to the late 20th century, which was around the same time period when video games experienced a surge in popularity and consumption. While the study does not directly examine video games, it highlights the importance of population-level indicators when evaluating educational strength across Europe. [3]

Although prior studies have explored the relationship between video games and academic outcomes, existing research has rarely investigated whether the growth of the video game industry itself is associated with changes in educational performance at a population level. This project addresses this gap by combining large-scale video game industry data with international PISA performance scores across European countries between 2003 and 2018. By analyzing trends over time, this study provides an informed, broader perspective on how changes in video game popularity may relate to educational outcomes.

## Hypothesis


Hypothesis: Increased video game sales per capita in Europe is strongly associated with lower average PISA scores among the population between 2003 and 2018.

The Programme for International Student Assessment (PISA) is an international standardized assessment conducted by the OECD that measures students’ literacy in reading, mathematics, and science, with an average score of approximately 500 across participating countries.

As stated in the background, due to the rise of video games in the mid twentieth century, countries were quick to adopt them, allowing games to become part of their culture and entertainment. More developed countries, such as Europe, were able to achieve this at a much faster rate. During that same time period in Europe, there was also a decline in educational performance as measured by the PISA Scores of 15 year-old students. We suggested that the rise of video games/video game sales is associated with lower educational performance in this age group, since video games can have varying impacts on students by taking away from study time and creating distractions. This does not explain a cause-and-effect relationship, but rather an association based on observations, which will be explored with supporting data.

### Second Analysis: PISA Score Performance in Europe 

The following explains ths Exploratory Data Analysis on Pisa Score Performace of student over 15 years old in categories such as reading, math, and science. It looks at the students performance specifically in Europe. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

RAW_PATH = "data/00-raw/OECD PISA data.csv"
pisa_ma = pd.read_csv(RAW_PATH)

if "index" in pisa_ma.columns:
    pisa_ma = pisa_ma.drop(columns=["index"])

europe = [
    'AUT','BEL','CZE','DNK','FIN','FRA','DEU','GRC','HUN','ISL',
    'IRL','ITA','LUX','NLD','NOR','POL','PRT','SVK','ESP','SWE',
    'CHE','GBR','LVA','EST','SVN','LTU'
]

#checking that the PISA Score is in Europe and ranges from (2000-2015)
pisa_eu = pisa_ma[
    (pisa_ma["LOCATION"].isin(europe)) &
    (pisa_ma["TIME"] >= 2000) &
    (pisa_ma["TIME"] <= 2015) &
    (pisa_ma["INDICATOR"] == "PISAMATH")
].copy()

#cleaned data
print("Filtered dataset shape:", pisa_eu.shape)
print(f"Missing values in 'Value' column: {pisa_eu['Value'].isnull().sum()}")

display(pisa_eu.head())
```

This section involves preparing the OECD PISA dataset for an exploratory analysis. The data is cleaned to include only European countries with observations between 2000 and 2015; to maintain consistency of geography and time. The focus for this analysis is on student mathematics performance using the PISA_MATH indicator to represent the average mathematics scores for all PISA assessed students. The dataset is then reviewed for missing values within the Value column (the column containing the average PISA score). Confirming that the value column does not contain any missing observations, allows the subsequent statistical analysis and visualization to be completed on a clean and reliable dataset.

```python
print("Mean:", pisa_eu["Value"].mean())
print("Median:", pisa_eu["Value"].median())
print("Skewness:", pisa_eu["Value"].skew())

#created histogram to display a visual exploring areas of interest 
plt.figure(figsize=(10, 5))
sns.histplot(pisa_eu["Value"], kde=True, bins=20)

plt.axvline(pisa_eu["Value"].median(), linestyle="--",
            label=f"Median: {pisa_eu['Value'].median():.2f}")
plt.axvline(pisa_eu["Value"].mean(), linestyle=":",
            label=f"Mean: {pisa_eu['Value'].mean():.2f}")

plt.title("Distribution of PISA Math Scores in Europe (2000–2015)", fontsize=14)
plt.xlabel("Math Score")
plt.ylabel("Count")
plt.legend()
plt.tight_layout()
plt.show()
```

Prior to analyzing specific countries, we will analyze how PISA math scores are distributed within our filtered data set. The distribution is shown in a histogram that includes a Kernel Density Estimate (KDE). Vertical reference lines for the mean and median are added to aid in evaluating if the distribution is symmetric. We will calculate a skewness value to evaluate if the distribution is skewed or if it is symmetrical. Once you understand the distribution of your scores, you can better decide which measure of central tendency to use to compare countries. If the distribution is slightly skewed, then the median may be a better measure than the mean because the median is less affected by extreme values.

```python
#listed the data in order from location and median values
order = pisa_eu.groupby("LOCATION")["Value"].median().sort_values(ascending=False).index
print("Top 5 by median:")
display(pisa_eu.groupby("LOCATION")["Value"].median().sort_values(ascending=False).head(5))
print("Bottom 5 by median:")
display(pisa_eu.groupby("LOCATION")["Value"].median().sort_values(ascending=False).tail(5))

#created a boxplot to display the variables of interests
plt.figure(figsize=(15, 7))
sns.boxplot(data=pisa_eu, x="LOCATION", y="Value", order=order)
plt.xticks(rotation=45)
plt.title("PISA Math Score Distribution by Country (Sorted by Median), 2000–2015", fontsize=16)
plt.ylabel("Score")
plt.xlabel("Country Code")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

To look at how well schools perform when compared across Europe, a boxplot was created to help view how math scores were distributed for each school. The countries were placed from highest to lowest median PISA score so the best performing countries are viewed as first on the boxplot. A box plot shows you the median score, the IQR and any possible outliers. Therefore, we can use this tool to evaluate not just the median of each country but the spread or variation of students' results. Countries with both a high median and narrow IQR will likely have good educational systems that produce both better overall and more consistently performing students.

```python
#looked at the top median values from the different countries in Europe
medians = pisa_eu.groupby("LOCATION")["Value"].median().sort_values(ascending=False)
top_3 = medians.head(3).index.tolist()
bottom_3 = medians.tail(3).index.tolist()

top_bottom = top_3 + bottom_3
print("Top 3 countries:", top_3)
print("Bottom 3 countries:", bottom_3)

trend_data = pisa_eu[pisa_eu["LOCATION"].isin(top_bottom)].copy()
trend_mean = trend_data.groupby(["LOCATION", "TIME"], as_index=False)["Value"].mean()

#created line plot exploring areas of interest
plt.figure(figsize=(12, 6))
sns.lineplot(data=trend_mean, x="TIME", y="Value", hue="LOCATION", marker="o", linewidth=2.5)
plt.title("Performance Trends: Top vs. Bottom Performing Countries (2000–2015)", fontsize=14)
plt.ylabel("Average Math Score")
plt.xlabel("Year")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

Although boxplots summarize performance across the whole duration of the time frame, they can also mask changes as time progresses. Line plots are used to visualize how the average math scores have changed per year, based on the performance of select countries. Only the top three and bottom three countries from the previous step, which were ranked by their median scores, will be analyzed. An average is calculated for each year, using the aggregated data by country and year to show the trend of each country's scores. By looking at the performance trends, it will be possible to determine if the average performance has remained consistent, improved, or decreased over the course of time. The trends observed in this type of analysis may represent changes in education as a whole; however, identifying specific causative factors would require additional context.

```python
#making sure there are no outliers
Q1 = pisa_eu["Value"].quantile(0.25)
Q3 = pisa_eu["Value"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = pisa_eu[(pisa_eu["Value"] < lower) | (pisa_eu["Value"] > upper)].copy()

print(f"Identified {len(outliers)} statistical outliers using IQR.")
print("Low outliers:", (outliers["Value"] < lower).sum())
print("High outliers:", (outliers["Value"] > upper).sum())

if len(outliers) > 0:
    display(outliers[["LOCATION", "TIME", "Value"]].sort_values(by="Value"))
    print("Countries with the most outliers:")
    display(outliers.groupby("LOCATION").size().sort_values(ascending=False).head(10))
else:
    print("No outliers detected in the filtered dataset.")
```

To better understand what appeared to be unusual in your data set, an Interquartile Range (IQR), is calculated to determine whether there are any outlier values in the PISA math test scores. Statistical outliers would include all scores that are less than Q1 minus 1.5 times IQR; and greater than Q3 plus 1.5 times IQR. Identifying these outliers can help you to see when a country has had a year of performance that was much different than expected for that country. This identification also gives you examples of things to look at more closely. Outliers may indicate actual improvements in education or, it could be due to either reporting errors or differences in methods that were used by each country in their PISA assessments.

### Combined Analysis 
The following analysis will merge both datasets in order to illustrate the relationship and possible correlation of both, helping us answer our research question.

First, we will get the average PISA scores across European countries by year.

```python
pisa_yearly = (
    pisa_eu
        .groupby("TIME", as_index=False)["Value"]
        .mean()
        .rename(columns={"TIME": "Year", "Value": "Avg PISA"})
)

pisa_yearly.head()
```

Now we will combine this data with the yearly video game sales to form our final analysis dataset.

```python
combined = yearly_sales.merge(
    pisa_yearly,
    on="Year",
    how="inner"
)

combined.head()
```

We will visualize the data using a plot.

```python
fig, ax1 = plt.subplots()

color1 = "tab:blue"
ax1.plot(combined["Year"], combined["EU Sales"], color = color1)
ax1.set_xlabel("Year")
ax1.set_ylabel("EU Sales (Millions)", color = color1)

color2 = "tab:orange"
ax2 = ax1.twinx()
ax2.plot(combined["Year"], combined["Avg PISA"], color = color2)
ax2.set_ylabel("Average PISA Score", color = color2)

plt.title("EU Video Game Sales vs Average PISA Scores")
plt.xticks(combined["Year"])
plt.show()
```

The plot displays trends in cumulative EU video game sales and average PISA scores across assessment years (2003–2015). EU video game sales (left axis, blue line) show an overall increase from 2003 to a peak in 2009, followed by a noticeable decline through 2015. In contrast, average PISA scores (right axis, orange line) show a gradual downward trend over the same period.

Visually, the two variables appear to move in opposite directions during certain intervals, particularly between 2003 and 2009, when sales increase while PISA scores decline slightly. However, after 2009, both variables decrease. This mixed pattern suggests that the relationship between the two variables is not consistently inverse over time.

While this plot is useful for comparing trends and observing how both variables evolve across years, it does not directly measure statistical association. Instead, we will use a scatterplot with a regression line to visualize the correlation, as it directly plots video game sales against PISA scores and allows for clearer evaluation of linear association.

```python
plt.figure()

plt.scatter(combined["EU Sales"], combined["Avg PISA"], color = "tab:blue")

m, b = np.polyfit(combined["EU Sales"], combined["Avg PISA"], 1)
plt.plot(combined["EU Sales"], m*combined["EU Sales"] + b, color = "tab:green")

plt.xlabel("EU Video Game Sales (Millions)")
plt.ylabel("Average PISA Score")
plt.title("Relationship Between EU Video Game Sales and PISA Scores")

plt.show()
```

The scatterplot demonstrates the relationship between total EU video game sales (in millions) and the average PISA score across corresponding assessment years. Each point represents one PISA testing year matched with total European video game sales in that same year.

The plotted points show a slight upward trend, as indicated by the positively sloped regression line. This suggests that higher levels of video game sales are associated with slightly higher average PISA scores during the observed years.

However, the points are relatively dispersed and do not follow a strong linear pattern. Additionally, the small number of observations (PISA is administered every three years) limits the strength of visual conclusions that can be drawn from the scatterplot.

```python
correlation = combined["EU Sales"].corr(combined["Avg PISA"])
print("Correlation:", correlation)
```

The calculated Pearson correlation coefficient is r = 0.242. This indicates a weak positive linear relationship, where as EU video game sales increase, average PISA scores tend to slightly increase. The relationship is weak and not strongly predictive.
Importantly, this result does not support the original hypothesis that increased video game sales are strongly associated with lower PISA scores. Instead, the data actually suggests a slight positive association.

```python
import statsmodels.api as sm

X = combined["EU Sales"]
y = combined["Avg PISA"]

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

print(model.summary())
```

A simple linear regression was conducted to examine whether cumulative EU video game sales predict average PISA scores. ## Discussion and Conclusion


This project sought to determine whether increases in video game sales in Europe are associated with lower educational performance as measured by PISA scores. Through combining large-scale video game industry data with international PISA performance scores across European countries between 2003 and 2015, we further expanded on past research, investigating whether the growth of the video game industry itself is associated with changes in educational performance at a population level. We initially hypothesized that increased video game sales per capita in Europe are strongly associated with lower average PISA scores among the population, as video games might be distracting for students and take time away from studying. 

Graphical visualizations of the time trends showed that EU video game sales increased between 2003 and 2009 before declining by 2015, while average PISA scores showed an overall gradual decrease over the same time period. Although these trends occasionally appeared to move in opposite directions, the pattern was not consistent across all years. We then used correlation and regression analysis to further examine the relationship between the two variables. The Pearson correlation coefficient was approximately 0.24, indicating a weak positive relationship between EU video game sales and PISA scores. The OLS regression produced a coefficient of 0.0138, meaning increases in video game sales were associated with a very small increase in PISA scores. However, the relationship was not statistically significant (p = 0.695), and the model explained only 5.8% of the variation in PISA scores (R² = 0.058). Overall, these results do not support the hypothesis that higher video game sales are associated with lower PISA performance.

There are several limitations affecting the interpretation of these findings. The biggest limitation is the small sample size, as only five observations remained after merging the datasets since PISA assessments only happen once every 3 years. The PISA dataset we used also only accounts for formally enrolled students, which does not represent all 15-year-olds accurately, while the Video Game Sales dataset only includes popular games with over 100,000 sales, excluding smaller indie games. Additionally, the analysis relies solely on combined regional data, which does not reflect individual student behavior or account for other factors that influence educational outcomes, like economic conditions, educational policies, or access to technology. 

Future research could be conducted to improve this analysis by breaking the data down by country, adding more relevant variables, and increasing the number of observations over time for a more detailed and accurate analysis. Overall, the results suggest that there is no strong statistical evidence linking higher video game sales to lower educational performance among students in Europe, and that the relationship between gaming and education is more likely to be influenced by other, broader factors.
