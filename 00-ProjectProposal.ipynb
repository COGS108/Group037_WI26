## Research Question

What is the association between the volume of video games released and education performance among the population in Europe during the 20th century?


## Background and Prior Work

The interactive media known as video games emerged in the mid 20th century and expanded rapidly during the 1980s and 1990s with the massive rise of home consoles and personal computing. By the end of the 20th century, tens of thousands of video games have been released globally, and millions of video games sold worldwide. Video games are not just a source of entertainment, but they are also a medium which has significantly shaped culture, technological literacy and society. Iconic characters such as Mario, Pikachu, Sonic, and Steve have become globally recognized symbols, reflecting the widespread cultural influence of video games.

With the critical success and acclaim of video games, their consumption worldwide has grown dramatically and continues to rise even today. Such consumption raises questions about the possible effects of video games, considering that it has become part of daily life for many individuals on a global scale. In Europe, accessibility to video games increased greatly throughout the 20th century, such that gaming became a common form of entertainment across different age groups and socioeconomic backgrounds. The widespread availability and impact of such forms of entertainment provide an opportunity to examine how varying levels of video game consumption correspond with educational performance indicators, such as literacy rates, standardized test scores, and broader academic achievement across European populations.

This growing interest in video games and its possible effects on education has prompted researchers and scholars. Previous studies have explored many distinct ways in which video games can affect an individual particularly when it comes to their education, including aspects of education such as learning, motivation, skill development and more. With this, and by situating it within the context of education in Europe, we can discover from previous work how historical patterns of video game consumption correspond with broader trends in academic achievement and literacy.

Prior research has examined a possible relationship between video games and academic performance in the settings of education. Merino-Campos et al. (2023) identified that factors such as accessibility, enjoyment, and technological competence significantly influenced students’ acceptance of video games as learning tools, and that this acceptance was associated with improved academic performance in physical education contexts. This work suggests that video game engagement can have measurable educational effects, supporting further investigation into broader relationships between video game consumption and educational outcomes. <sup><a href="https://link.springer.com/article/10.1007/s10639-022-11295-y">[1]</a></sup>

Other studies show that not only the relationship of learning tools with video games, but also the social and behavioral effects that come with such media. Squire (2003) observed in their research that while very often this form of media was ignored by educators, video games have become a relevant resource for the development and improvement of cultural competences, as well as  social and cognitive skills by secondary school students in the context of physical education. They draw on theories revolving around motivation, which they conclude that deeper learning and engagement can be produced by the key elements of video games such as the challenge, feedback, and other aspects. <sup><a href="https://boardwork317.wordpress.com/wp-content/uploads/2016/03/video-games-in-education.pdf">[2]</a></sup>

Finally, de Wit, Ferencz, and Rumbley (2018) focus their research on international mobility in Europe at a population level. They examine international mobility as a major indicator of educational strength, ability, accessibility, and quality across Europe. This research dates to the late 20th century, which was about the same period where video games rose in popularity and consumption. <sup><a href="https://www.taylorfrancis.com/chapters/edit/10.4324/9781315160870-4/international-student-mobility-hans-de-wit-irina-ferencz-laura-rumbley">[3]</a></sup>


## Hypothesis


Hypothesis: Increased video game sales per capita in Europe is strongly associated with lower average PISA scores among the population within the 20th Century. 

As stated in the background, because video games became the “new trend” in the twentieth century, countries were quick to adopt it; allowing games to become part of its culture and entertainment. More developed countries such as Europe were able to consume that at a much faster rate. During that same time period in Europe, there was also a decline in educational performance (PISA Scores) ranging from literacy score to test rates. We suggested that the rise of video games/video games sales is associated with lower educational performance since video games can have varying impacts on different people as it takes away time from studying and creates distractions. This does not explain a cause and effect relationship, rather an association based on observations which will be explored with supporting data.

## Ethics 


Overall statement: This project is based on aggregated, publicly available databases and therefore does not contain any personally identifiable information (PII). However, it is possible that this data could be biased through collection biases resulting from differences in how consistently video game sales have been reported across countries and over time, especially in earlier time frames; additionally, the measures of educational performance are likely to be impacted by broader systemic factors that do not directly relate to video game consumption, such as the country’s overall economic development or its academic policies.
To minimize these limitations, this project will not make any causal statements about the relationship between video games and educational outcomes. The project will state all limitations and will present the results using graphical representations and summary statistics that will allow for transparency and show the uncertainty within the data sets. Additionally, any observed correlations should not be interpreted as evidence of individual-level effects, and no policy recommendations can be made based on the observed correlations without additional supportive evidence.
