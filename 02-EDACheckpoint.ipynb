## Dataset #1: Yearly Video Game Sales By Genre
Link to dataset: https://public.tableau.com/app/profile/yue.wang7536/viz/VideoGameSalesKaggle_17350173664110/VideoGameSales 

#### Description 
This dataset contains data on video games with sales greater than 100,000 copies, collected by a scrape of vgchartz.com. It includes the following information: ranking of overall sales, name of game, platform, release year, genre, publisher, and sales of video games. The metric that is used within this dataset is per million copies sold, where a copy represents one singular video game purchased. This metric is categorized in regions; in North America(NA_Sales), Europe(EU_Sales), Japan(JP_Sales), Other(Other_Sales), and Global(Global_Sales), where Global_Sales represents the sum of all regional sales. Larger numbers of copies sold indicate the success and popularity of a video game. We will specifically be using the data on video game Sales by Year and Genre, which tracks and displays the trend of sales by genre from 1980-2020 through the use of a stacked area chart. In regards to our analysis, we will also be specifically only looking at the data on video game sales from 2000-2015 in the EU region, which is represented in the dataset as Zone Sales.

#### Concerns regarding Dataset 
- Given the sales of millions of video games per region and globally, it is crucial to note that a concern can be that each region contains different population numbers in their respective areas. For instance, North America has a vastly different population to other regions like Europe or Japan. With larger populations comes larger potential consumer markets, meaning that the release and sale of video games can weigh more on market size than actual popularity of a video game. Therefore, using a dataset which describes the raw total sales of video games in all time can introduce bias given the different populations of each region.

- It is essential to consider the different regulations or laws the different regions have set for the consumption of video games, which can influence the release and sale of video games. For instance, the majority of regions have regulatory organizations that mainly enforce ratings, advertising guidelines, and online privacy principles for video game users. It is important to understand that while the majority of regions have such regulation, some regions have more strict laws than others, as well as some countries completely banning the consumption of video games. This explains that the regulations per region can effectively control the release and sale of video games.

- It is also important to note that there may be a selection bias as only games with over 100,000 copies sold are included, which favors larger, more popular game companies and franchises. This could mean an underrepresentation of indie and less popular games.

## Dataset #2: PISA Performance of students from different countries

Link to dataset: https://www.kaggle.com/datasets/thedevastator/pisa-performance-scores-by-country

#### Description
This dataset is a result of the Organisation for Economic Co-operation and Development (OECD)'s Programme for International Student Assessment (PISA), an international study assessing the educational attainment of 15-year-old students. PISA evaluates whether students can apply their knowledge in practical, real-life contexts.

The dataset includes the average PISA performance scores for reading, mathematics, and science. These are scale scores (not percentages) and are standardized so that the OECD average is approximately 500. A difference of 30-40 points is generally interpreted as roughly one year of schooling.

Reading literacy measures the ability to understand and use written texts. Mathematics literacy measures the ability to apply mathematical knowledge in various contexts. Scientific literacy measures the ability to apply scientific knowledge and evidence to explain phenomena and draw conclusions.

Additional categorical variables include the country's alpha-3 code, the domain assessed (reading, math, science), subject group (male, female, or total), year of assessment, and the mean score value. Because PISA rotates its primary domain each cycle, measurement precision can vary across years.

#### Concerns regarding Dataset
- PISA only samples 15-year-old students who are enrolled in school, excluding non-enrolled adolescents.
- Cultural, language, and curriculum differences may influence how students interpret assessment items.
- National mean scores may mask within-country inequality across regions and socio-economic groups.
- Changes in testing frameworks, sampling strategies, or participation over time may affect trend validity.

```python
pisa = pisa_ma[pisa_ma['INDICATOR'].str.contains('PISA', case=False, na=False)]
europe = ['FRA','DEU','ITA','ESP','NLD','SWE','NOR','FIN','DNK','POL','PRT','GRC','CZE','AUT','BEL','IRL','HUN','CHE','SVK','SVN','EST','LVA','LTU']
math_scores = pisa[(pisa['LOCATION'].isin(europe)) & (pisa['SUBJECT']=='MATH')]
Q1 = math_scores['Value'].quantile(0.25)
Q3 = math_scores['Value'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
outliers = math_scores[(math_scores['Value'] < lower) | (math_scores['Value'] > upper)]
outliers[['LOCATION','TIME','Value']].sort_values('Value')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

europe = ['AUT', 'BEL', 'CZE', 'DNK', 'FIN', 'FRA', 'DEU', 'GRC', 'HUN', 'ISL',
          'IRL', 'ITA', 'LUX', 'NLD', 'NOR', 'POL', 'PRT', 'SVK', 'ESP', 'SWE',
          'CHE', 'GBR', 'LVA', 'EST', 'SVN', 'LTU']

pisa_ma = pisa_ma[(pisa_ma['TIME'] >= 2000) & (pisa_ma['TIME'] <= 2015)]

europe_scores = pisa_ma[pisa_ma["LOCATION"].isin(europe)]
plt.figure(figsize=(14,6))
sns.boxplot(data=europe_scores, x="LOCATION", y="Value")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

There are no suspicious entries or outliers as the range of data is reasonable. Since there was no missing data and extraneous data, there was no need to remove a speific data from the dataset. #### Analysis of Yearly Video Game Sales by Genre Dataset

```python
vg_sales = pd.read_csv("data/02-processed/clean_vg_sales.csv")
vg_sales
```

```python
yearly_sales = (
    vg_sales
        .groupby("Year", as_index = False)["Zone Sales"]
    .sum()
)
yearly_sales
```

```python
plt.figure()
plt.plot(yearly_sales["Year"], yearly_sales["Zone Sales"])
plt.xlabel("Year")
plt.ylabel("Total EU Sales (Millions)")
plt.title("Total Video Game Sales in Europe (2000-2015)")
plt.xticks(yearly_sales["Year"], rotation = 45)
plt.ylim(0,200)
plt.show()
```

```python
yearly_sales.describe()
```

The trend line and summary statistics of yearly video game sales from 2000–2015 provide insight into overall market performance during this period. The data shows that peak yearly sales occurred in 2009, reaching a maximum of 191.59 million units, which likely represents the height of the video game market within this timeframe. In contrast, the lowest recorded sales were 52.75 million units, indicating that the market experienced substantial changes in performance across the years. This variability is further supported by the standard deviation of 37.71 million, suggesting notable fluctuations in yearly sales rather than consistent growth. Additionally, the interquartile range (25%–75%) shows that most yearly sales fell between approximately 106 million and 162 million units, representing the typical performance range of the video game market between 2000 and 2015. Overall, these statistics highlight both periods of strong expansion and noticeable decline within the industry.

```python
sns.boxplot(data = yearly_sales['Zone Sales']);
```

Based on the boxplot, there are no significant outliers in yearly sales. This indicates that video game sales from 2000 to 2015 changed gradually over time, without extreme spikes or declines in any given year.

```python
years = range(1, 16)
sales_change = [42, 15, -5, 4, 15, 8, 32, 24, 8, -15, -9, -48, 8, 0, -28]

plt.figure(figsize=(10, 2)) 


plt.plot(years, sales_change, marker='o', color='tab:orange', linewidth=2, label='Yearly Change')


plt.axhline(0, color='gray', linewidth=1.5, linestyle='-') 


plt.title('Year-to-Year Change in Video Game Sales', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Change in Sales (%)', fontsize=12)
plt.xticks(years) # Ensures every year is labeled clearly
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

The following plot shows the year-to-year trend in total video game sales from 2000 to 2015, along with a horizontal line at 0 indicating whether sales increased or decreased compared to the previous year. From the plot, sales in the early 2000s increased rapidly before slightly decreasing around 2003. Sales then increased again throughout the remainder of the decade, reflecting a period of rapid growth in the video game market. The transition into the 2010s shows a significant downturn, as sales decreased between 2010 and 2012 by approximately 15–40 million units compared to previous years. In 2013, sales experienced a noticeable increase, followed by another decline beginning in 2014 and continuing toward the end of the observed period. Overall, the changes observed from 2010 onward suggest slower growth in video game sales compared to the previous decade.

To measure the overall trend of video game sales in Europe, we aggregated EU sales across all genres for each year from 2000-2015. This gave us a cumulative yearly sales metric that represents total market consumption. The trend shows significant growth in the early to mid 2000s, with a peak around the late 2000s. This cumulative measure will be compared with PISA scores of 15 year-old European students in corresponding years.
